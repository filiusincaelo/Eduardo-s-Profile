## Age Distributions and Totals

```python
# Importing libraries to read csv data and plot
import pandas as pd
import matplotlib.pyplot as plt

data_ages = pd.read_csv("datacensus.csv", sep=";", nrows=22, usecols=[0, 1, 2, 3])
```

```python
data_ages.head(22)
```

Note: Table with ages information

```python
# Extracting data
age_ranges = data_ages['Description']
totals = data_ages['Total']

# Creating the bar plot
plt.figure(figsize=(10, 6))
plt.bar(age_ranges, totals, color='skyblue')
plt.xlabel('Age Ranges')
plt.ylabel('Total Population in Millions')
plt.title('Population Distribution by Age Ranges in Canada')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from being cut off
plt.show()
```

Note: Most of the population ranged in 2015 from 20 to 59 years

```python
# reading data
data_ages_percent = pd.read_csv("datacensus.csv", sep=";", nrows=4, usecols=[0, 1, 2, 3], skiprows=25)

data_ages_percent
```

Note: Table with percentages

```python
# Extracting data
age_groups = data_ages_percent['Description']
percent_total = data_ages_percent['Total']

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percent_total, labels=age_groups, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen','lightyellow'])
plt.title('Percentage by Age Groups in Canada', pad=20)  # Adjust pad value as needed
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```

## Income Comparison between 2016 and 2021

```python
data_income = pd.read_csv("datacensus.csv", sep=";", nrows=4, usecols=[0, 1], skiprows=60)

data_income
```

```python
# Data
categories = ['Median Income FT workers', 'Average Income']
income_2016 = [34204, 47487]
income_2021 = [40500, 54630]

# Bar positions
x = range(len(categories))

# Bar width
width = 0.35  

# Create bars for 2016 data
plt.bar(x, income_2016, width, label='2016', color='skyblue')

# Create bars for 2021 data
plt.bar([i + width for i in x], income_2021, width, label='2021', color='lightcoral')

# Labels and title
plt.xlabel('Income Category')
plt.ylabel('Income')
plt.title('Income Comparison between 2016 and 2021')
plt.xticks([i + width/2 for i in x], categories)
plt.legend()

# Show plot
plt.show()
```

Note: The Income varies between occupations, there is a tendency that male earn more money than female

## Latin American Canadian Population in 2016

```python
# reading data
data_latin_american = pd.read_csv("datacensus.csv", sep=";", nrows=26, usecols=range(0, 5), skiprows=67)

data_latin_american
```

```python
# Group data by Province and Category and sum the values
grouped_df = data_latin_american.groupby(['Province', 'Category']).sum().reset_index()

# Pivot the DataFrame to have 'Category' as columns
pivot_df = grouped_df.pivot(index='Province', columns='Category', values='Total')

# Plotting the bar chart
pivot_df.plot(kind='bar', stacked=False, figsize=(12, 8))
plt.title('Total by Category and Province')
plt.ylabel('Total')
plt.xlabel('Province')
plt.legend(title='Category')
plt.xticks(rotation=90)
plt.show()
```

Note: The information provided is in thousands. The blue bar, representing 'Latin American,' likely denotes the population predominantly composed of individuals born in Latin American countries, thus comprising Latin American immigrants. On the other hand, the orange bar, labeled 'Latin American origins,' encompasses individuals who identify as being of Latin American descent, regardless of their place of birth. This distinction helps differentiate between immigrants and those with ancestral ties to Latin America

## Reported Diabetes in Canada 2016 - 2019

```python
# reading data
data_diabetes = pd.read_csv("datacensus.csv", sep=";", nrows=4, usecols=range(0, 3), skiprows=101)

data_diabetes
```

Note: This table includes both sexes, some information of other ages was not published due to its lack of reliability or erros

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(data_diabetes['Diabetes'], data_diabetes['Percentage'], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Diabetes Percentage')
plt.title('Diabetes Percentage by Age Group')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Note: It can be seen that as people get older, more likely is to have diabetes

## Consumer Preferences in Canada, survey 2021 (187,000 surveys)

```python
# reading data
data_consumer_preferences = pd.read_csv("datacensus.csv", sep=";", nrows=9, usecols=[0, 1], skiprows=109)

data_consumer_preferences
```

Note: Data retrieved from Statista

```python
# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data_consumer_preferences['Description'], data_consumer_preferences['Percentage'], color='orange')
plt.xlabel('Percentage')
plt.ylabel('Statement')
plt.title('Consumer perspective of natural and organic food and drink in Canada in 2021')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest percentage at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Note: Almost 4 of out 10 Canadians will associate an organic product to be a healthy product, yet only about 20% is willing to pay more for organic products.
